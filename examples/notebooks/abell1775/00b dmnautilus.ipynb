```bash
cd 13510
```

```bash
punlearn dmnautilus
plist dmnautilus
/bin/rm -rf dmnautilus
mkdir dmnautilus
```

```bash
punlearn dmnautilus
time dmnautilus \
  infile=img.fits \
  outfile=dmnautilus/img.abin \
  snr=10 \
  method=0 \
  outmask=dmnautilus/img.map \
  outsnr=dmnautilus/img.snr \
  outarea=dmnautilus/img.area \
  mode=h clob+
```

```bash
ds9 -geometry 933x576 \
  -view colorbar no -view info no -view panner no -view magnifier no \
  dmnautilus/img.abin -scale log \
  -cmap load sunset-in-atlantis.sao \
  -region dmnautilus/img.map"[region]" \
  -region select all -region color black -region select none \
  -block to 2 \
  dmnautilus/img.map -scale linear \
  -region select all -region color black -region select none \
  -cmap load $ASCDS_CONTRIB/data/16_ramps.lut \
  -block to 1 -zoom to 0.5 \
  -tile column \
  -saveimage png dmnautilus/img.abin.png -quit

display < dmnautilus/img.abin.png
```

```bash
dmstat dmnautilus/img.snr sig- med- cen- | egrep 'min|max'
```
